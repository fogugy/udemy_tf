```python
import numpy as np
import tensorflow as tf
import tensorflow.keras as kr
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
N = 1000
X = np.random.random((N, 2)) * 6 - 3
Y = np.cos(2*X[:,0]) + np.cos(3*X[:,1])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], Y, color='blue');
```

```python
model = kr.models.Sequential([
    kr.layers.Dense(2, activation=tf.nn.leaky_relu),
    kr.layers.Dense(120, activation=tf.nn.leaky_relu),
    kr.layers.Dense(1)
])
```

```python
model.compile(optimizer=kr.optimizers.Adam(0.005), loss='mse')
```

```python
r = model.fit(X, Y, epochs=100)
```

```python
plt.plot(r.history['loss'], label='loss');
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], Y, color='blue')
ax.scatter(X[:,0], X[:,1], model.predict(X), color='green');
```
